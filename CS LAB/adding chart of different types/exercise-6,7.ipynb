# II. Data Visualization

```python

import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("company_sales_data.csv")
df.head()
```

## 1.Matplotlib package and plotting different graphs


### Read Total profit of all months and show it using a line plot

```python

profitList = df ['total_profit'].tolist()
monthList  = df ['month_number'].tolist()
plt.plot(monthList, profitList)
plt.xlabel('Month')
plt.ylabel('Profit')
plt.title('profit per month')
plt.show()
```

### Get total profit of all months and show line plot with the following Style properties

```python
 
profitList = df ['total_profit'].tolist()
monthList  = df ['month_number'].tolist()
plt.plot(monthList, profitList, label = 'Profit data', color='orange', marker='o', markerfacecolor='black', linestyle=':', linewidth=2)
plt.xlabel('Month Number')
plt.ylabel('Profit')
plt.title('Sales data')
plt.legend()
plt.show()
```

###  Read all product sales data and show it  using a multiline plot

```python

monthList  = df ['month_number'].tolist()
faceCremSalesData   = df ['facecream'].tolist()
faceWashSalesData   = df ['facewash'].tolist()
toothPasteSalesData = df ['toothpaste'].tolist()
bathingsoapSalesData   = df ['bathingsoap'].tolist()
shampooSalesData   = df ['shampoo'].tolist()
moisturizerSalesData = df ['moisturizer'].tolist()

plt.plot(monthList, faceCremSalesData,   label = 'Facecream', marker='o', linewidth=2)
plt.plot(monthList, faceWashSalesData,   label = 'FaceWash',  marker='o', linewidth=2)
plt.plot(monthList, toothPasteSalesData, label = 'ToothPaste', marker='o', linewidth=2)
plt.plot(monthList, bathingsoapSalesData, label = 'Bathingsoap', marker='o', linewidth=2)
plt.plot(monthList, shampooSalesData, label = 'Shampoo', marker='o', linewidth=2)
plt.plot(monthList, moisturizerSalesData, label = 'Moisturizer', marker='o', linewidth=2)

plt.xlabel('Month')
plt.ylabel('Sales units')
plt.legend(loc='upper left')
plt.title('Sales data')
plt.show()
```

### Read moisturizer sales data of each month and show it using a scatter plot

```python

monthList  = df ['month_number'].tolist()
moisturizerSalesData = df ['moisturizer'].tolist()
plt.scatter(monthList, moisturizerSalesData, label = 'moisturizer Sales data', color='red')
plt.xlabel('Month')
plt.ylabel('Number of units Sold')
plt.title('moisturizer Sales data')
plt.legend(loc = 'upper right')
plt.show()
```

### Read moisturizer and shampoo product sales data and show it using the bar chart

```python

monthList  = df ['month_number'].tolist()
faceCremSalesData   = df ['moisturizer'].tolist()
faceWashSalesData   = df ['shampoo'].tolist()

plt.bar([a-0.25 for a in monthList], faceCremSalesData, width= 0.25, label = 'moisturizer sales data', align='edge', color='black')
plt.bar([a+0.25 for a in monthList], faceWashSalesData, width= -0.25, label = 'facewash sales data', align='edge', color='orange')
plt.xlabel('Month')
plt.ylabel('Sales units')
plt.legend(loc='upper right')
plt.title('Sales data')
plt.title('moisturizer and shampoo sales data')
plt.show()
```

### Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python

profitList = df ['total_profit'].tolist()
plt.hist(profitList, edgecolor='black')
plt.xlabel('profit range')
plt.ylabel('Actual Profit')
plt.title('Profit data')
plt.show()
```

### Calculate total sale data for last year for each product and show it using a Pie chart

```python

monthList  = df ['month_number'].tolist()
labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesData   = [df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(), df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]
plt.axis("equal")
plt.pie(salesData, labels=labels, autopct='%1.1f%%')
plt.legend(loc='upper left')
plt.title('Sales data')
plt.show()
```

###  b)Adding text "Average chart of a company " x axis name , y axis name

```python
profitList = df ['total_profit'].tolist()
profitList = [i/6 for i in profitList]
monthList  = df ['month_number'].tolist()
plt.plot(monthList, profitList)
plt.xlabel('Month')
plt.ylabel('Average Profit')
plt.title('Average chart of a company')
plt.legend()
plt.show()
```
